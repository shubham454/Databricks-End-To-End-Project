## Dynamic capabilities

```python
dbutils.widgets.text("p_file_name","")
p_file_name = dbutils.widgets.get("p_file_name")
```

### Data Loading

```python
df = spark.readStream.format("cloudFiles")\
              .option("cloudFiles.format", "parquet")\
              .option("cloudFiles.schemaLocation", f"/Volumes/db_dlt_proj/ext_location/bronze_external_location/checkpoint_{p_file_name}")\
              .load(f"/Volumes/db_dlt_proj/ext_location/source_external_location/{p_file_name}/")
```

```python
df.writeStream\
    .format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"/Volumes/db_dlt_proj/ext_location/bronze_external_location/checkpoint_{p_file_name}")\
    .option("path", f"/Volumes/db_dlt_proj/ext_location/bronze_external_location/{p_file_name}")\
    .trigger(once=True)\
    .start()
```
