```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
df = spark.read.format("parquet")\
                .load("/Volumes/db_dlt_proj/ext_location/bronze_external_location/customers/")
df = df.drop("_rescued_data")
```

## Transformations

```python
# create domain column from the email filed
df = df.withColumn("domain", split(col("email"),"@").getItem(1))
```

```python
df = df.withColumn("full_name", concat(col("first_name"), lit(" "),col("last_name"))).drop("first_name","last_name")
# df.show()
```

# Data writing in delta table

```python
df.write.format("delta")\
        .mode("overwrite")\
          .save("/Volumes/db_dlt_proj/ext_location/silver_external_location/customers/")

df.write.format("delta")\
        .mode("overwrite")\
          .saveAsTable("db_dlt_proj.silver.customers")
```
