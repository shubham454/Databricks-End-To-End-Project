### Data Reading

```python
from pyspark.sql.functions import col, year, to_timestamp
df = spark.read.format('parquet')\
                .load("/Volumes/db_dlt_proj/ext_location/bronze_external_location/orders/")
```

```python
df = df.drop("_rescued_data")
df = df.withColumn("order_date",to_timestamp(col('order_date')))
df = df.withColumn("year", year(col('order_date')))
```

### Data writing to delta lake

```python
df.write.format("delta")\
  .mode("overwrite")\
    .save("/Volumes/db_dlt_proj/ext_location/silver_external_location/orders")


df.write.format("delta")\
        .mode("overwrite")\
          .saveAsTable("db_dlt_proj.silver.orders")
```
